```python
df = df.groupby(['title'], as_index=False).first()
df['genre'] = df['genre'].astype(str)
df = df[df['genre'] != 'nan']
df
```

```python
df['genre'] = [i.split(', ') for i in df['genre']]
```

```python
df['title'].nunique(), len(df)
```

```python
# Check the frequency of each category
categories = {}
for item in df['genre']:
    for category in item:
        if category in categories:
            categories[category] += 1
        else:
            categories[category] = 0
    
print("NUmber of categories: {}".format(len(categories)))
```

```python
plt.figure(figsize=(20,20))
plt.barh(list(categories.keys()), list(categories.values()))
```

```python
X = df['synopsis']
y = df['genre']
```

```python
import re
import string
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

english_stopwords = set(stopwords.words('english'))

def preprocess_text(data):
    data = [str(i).lower() for i in data]
    data = [i.replace(string.punctuation, ' ') for i in data]
    data = [re.sub(r'\W', ' ', i) for i in data]
    data = [re.sub(r'\s+', ' ', i) for i in data]
    data = [word_tokenize(i) for i in data]
    data = [[j for j in i if j not in english_stopwords] for i in data]
    data = [' '.join(i) for i in data]
    return data

X = preprocess_text(X)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(X)
```

```python
X = X.toarray().reshape(X.shape[0], 1, X.shape[1])
```

```python
X.shape
```

```python
from sklearn.preprocessing import MultiLabelBinarizer
```

```python
mlb = MultiLabelBinarizer()
y = mlb.fit_transform(y)
```

```python
mlb.classes_
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
X_train.shape
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout, LSTM
from keras.optimizers import Adam
```

```python
print("X_train: {}, y_train: {}".format(X_train.shape, y_train.shape))
```

```python
model = Sequential()
model.add(LSTM(128, input_shape=X_train[0].shape, activation='relu', return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(128, activation='relu'))
model.add(Dropout(0.2))

model.add(Dense(64, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(len(categories), activation='sigmoid'))

opt = Adam(lr=1e-3, decay=1e-5)
model.compile(loss='binary_crossentropy', optimizer=opt, metrics=['accuracy'])
model.summary()
```

```python
from keras.callbacks import EarlyStopping, ModelCheckpoint

callbacks = [
    EarlyStopping(monitor='accuracy', patience=3),
    ModelCheckpoint('checkpoint.h5')
]

model.fit(X_train, y_train, epochs=100, callbacks=callbacks)
```

```python
to_test = np.reshape(X_test[0], (1, X_test[0].shape[0], X_test[0].shape[1]))
predictions = model.predict(to_test)
predictions
```

```python
result = model.evaluate(X_test, y_test)
print(dict(zip(model.metrics_names, result)))
```

```python
def predict(text, THRESHOLD = 0.4):
    text = preprocess_text([text])[0]
    vectorized_text = vectorizer.transform([text])
    vectorized_text = vectorized_text.toarray().reshape(vectorized_text.shape[0], 1, vectorized_text.shape[1])
    prediction = model.predict(vectorized_text)[0]
    transformed_prediction = np.asarray([[1 if i > THRESHOLD else 0 for i in prediction]])
    categories = mlb.inverse_transform(transformed_prediction)[0]
    
    probabilities = [i for i in prediction if i > THRESHOLD]
    return dict(zip(categories, probabilities))
```
